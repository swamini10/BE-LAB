```python
 import matplotlib.pyplot as plt
```

```python
 def f(x):
    return (x + 3)**2
 def df(x):
    return 2 * (x + 3)
```

```python
 x_old = 2          # Starting point
 learning_rate = 0.1
 precision = 0.00001
 max_iterations = 1000
```

```python
 x_values = []
 y_values = []
 for i in range(max_iterations):
    gradient = df(x_old)
    x_new = x_old - learning_rate * gradient
    x_values.append(x_new)
    y_values.append(f(x_new))
    if abs(x_new - x_old) < precision:
        break
    x_old = x_new
 print(f"Local minima occurs at x = {x_new:.5f}")
 print(f"Minimum value of function y = {f(x_new):.5f}")
```

```python
 plt.plot(x_values, y_values, 'b-', label='Gradient Descent Path')
 plt.title("Gradient Descent Convergence")
 plt.xlabel("x")
 plt.ylabel("y = (x + 3)^2")
 plt.legend()
 plt.grid(True)
 plt.show()
```
